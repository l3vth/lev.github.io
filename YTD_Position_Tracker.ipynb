```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")
```

# Dates

```python
start_date = "2023-01-01"
end_date = "2023-12-12"
```

# Data Load

```python
tickers = ["UPRO","TMF","AVUV","VEA","AVDV","VWO","DGS","DBMF","KRBN","VTI","PDBC","VPU","BTC-USD"]
```

```python
wts = [0.363661811
,0.296892618
,0.201476832
,0.028044193
,0.02804104
,0.019744751
,0.018696607
,0.010184755
,0.009968619
,0.009965036
,0.006776746
,0.006546994
,0.071663063 ]
```

```python
yf.pdr_override()

price_data = pdr.get_data_yahoo(tickers,
                               start = start_date,
                               end = end_date)
```

```python
price_data = price_data['Adj Close']
```

```python
ret_data = price_data.pct_change()[1:]
```

```python
weighted_returns = (wts * ret_data)
```

```python
weighted_returns = weighted_returns.reset_index()
weighted_returns.head()
```

```python
port_ret = weighted_returns.sum(axis=1)
# axis =1 tells pandas we want to add
```

```python
cumulative_ret = ((port_ret + 1).cumprod()) - 1
```

# Graph Returns

```python
fig = go.Figure() # go object

#Date = weighted["Date"]
#benchmark = port_ret["VTI"]
vala = cumulative_ret
VTI = ((ret_data["VTI"] + 1).cumprod()) - 1


fig.add_trace(go.Scatter(x=weighted_returns["Date"],y=cumulative_ret, name='Valas Return',
                         line=dict(color='royalblue', width=3,)))
fig.add_trace(go.Scatter(x=weighted_returns["Date"], y=VTI, name='VTI',
                        line=dict(color='green', width=3,)))

# Edit the layout
fig.update_layout(title=f'Valas Portfolio {(cumulative_ret.tail(1).values *100).round(2)}% vs VTI {(VTI.tail(1).values *100).round(2)}% - 2023',
                   xaxis_title='Year',
                   yaxis_title='Cumulative Return', width=1000, height=800,)
fig.show()
#Saving Plot for Website
```
